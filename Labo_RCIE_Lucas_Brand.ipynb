```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
import random
```

```python
def openFile(file): #angle_or_duration=0):
    num, time, angle, duration, errorX, errorY = [],[],[],[],[],[]
    with open(file, 'r') as f:
        ligne = f.readline()
        ligne = f.readline()
        while(ligne):
            #ligne = f.readline()
            n, t, a, d, eX, eY = ligne.split(';')
            num.append(int(n))
            time.append(float(t))
            duration.append(float(d))
            angle.append(float(a))
            errorX.append(float(eX))
            errorY.append(float(eY)*float(eY))
            ligne=f.readline()
    data = {'num': num,
           'time': time,
           'angle': angle,
           'duration': duration,
           'errorX': errorX,
           'errorY': errorY}
    return data
```

```python
def afficheResult(data, angle_or_duration, title):
    if angle_or_duration == 'angle':
        plt.plot(data['angle'], data['errorY'], 'bo')
        plt.xlabel('angle')
        plt.title(title)
        
    if angle_or_duration == 'duration':
        plt.plot(data['duration'], data['errorY'], 'ro')
        plt.xlabel('duration')
        plt.title(title)
        
    plt.ylabel("errorY^2")
    plt.show()
```

```python
# A (10) : Variation angle avec dur√©e fixe
angle_10 = openFile('.\data\expePsychoResults_1_10.csv')
afficheResult(angle_10, 'angle', 'Angle variable et dur√©e fixe - 10 √©chantillons')

# A (100) : Variation angle avec dur√©e fixe
angle_100 = openFile('.\data\expePsychoResults_1_100.csv')
afficheResult(angle_100, 'angle', 'Angle variable et dur√©e fixe - 100 √©chantillons')

# D (10) : Variation dur√©e avec angle fixe
duration_10 = openFile('.\data\expePsychoResults_2_10.csv')
afficheResult(duration_10, 'duration', 'Angle fixe et dur√©e variable - 10 √©chantillons')

# D (100) Variation dur√©e avec angle fixe
duration_100 = openFile('.\data\expePsychoResults_2_100.csv')
afficheResult(duration_100, 'duration', 'Angle fixe et dur√©e variable - 100 √©chantillons')
```
